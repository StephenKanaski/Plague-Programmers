```python
import pandas as pd
import datetime
import numpy as np
```

```python
covid_data = "us-states.csv"
states = "state_populations.csv"
```

```python
covid = pd.read_csv(covid_data)
state_data = pd.read_csv(states)
```

```python
state_data = state_data.replace(',','', regex=True)
```

```python
state_data['2019'] = state_data['2019'].astype(float)
```

```python
covid.dtypes
```

```python
covid['date'] =  pd.to_datetime(covid['date'], format='%Y-%m-%d')
```

```python
covid.dtypes
```

```python
covid_filtered = covid.loc[covid["date"] == "2020-10-06", :]
```

```python
covid_filtered
```

```python
state_data
```

```python
covid_filtered
```

```python
covid_filter = covid_filtered[covid_filtered.state != ('Northern Mariana Islands', 'Virgin Islands')]
```

```python
covid_filter
```

```python
states = state_data.rename(columns={'State': 'state', '2019': 'population'})
```

```python
covid_cases = pd.merge(covid_filter, states, on='state')
```

```python
covid_cases
```

```python
covid_cases['cases per 100K'] = round((covid_cases.cases / covid_cases.population) * 100000)
```

```python
covid_cases
```

```python
covid_cases.to_csv('covid_cases.csv', index=False)
```
